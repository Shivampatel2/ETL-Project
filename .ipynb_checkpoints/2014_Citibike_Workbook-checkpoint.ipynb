```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('resources/2014_citibike_data.csv')
df.head(10)
```

```python
new_df = df.drop(['start station longitude', 'start station latitude', 'end station longitude', 'end station latitude'], axis=1)
new_df.rename(columns={'tripduration': 'Trip Duration'}, inplace=True)
new_df.rename(columns={'starttime': 'Start Time'}, inplace=True)
new_df.rename(columns={'stoptime': 'Stop Time'}, inplace=True)
new_df.rename(columns={'bikeid': 'Bike ID'}, inplace=True)
new_df.rename(columns={'usertype': 'User Type'}, inplace=True)
new_df.rename(columns={'birth year': 'Birth Year'}, inplace=True)
new_df.rename(columns={'gender': 'Gender'}, inplace=True)
new_df.rename(columns={'start station name': 'Start Station Name'}, inplace=True)
new_df.rename(columns={'end station name': 'End Station Name'}, inplace=True)
new_df.rename(columns={'start station id': 'Start Station ID'}, inplace=True)
new_df.rename(columns={'end station id': 'Start Station ID'}, inplace=True)
new_df.tail(10)
```
