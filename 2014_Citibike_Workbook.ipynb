```python
import pandas as pd
import numpy as np
```

```python
df_2014 = pd.read_csv('resources/2014_citibike_data.csv')
df_2014.head(10)
```

```python
# remove lat/lng columns and renamed others
new_df_2014 = df_2014.drop(['start station longitude', 'start station latitude', 'end station longitude', 'end station latitude'], axis=1)
new_df_2014.rename(columns={'tripduration': 'Trip_Duration'}, inplace=True)
new_df_2014.rename(columns={'starttime': 'Start_Time'}, inplace=True)
new_df_2014.rename(columns={'stoptime': 'Stop_Time'}, inplace=True)
new_df_2014.rename(columns={'bikeid': 'Bike_ID'}, inplace=True)
new_df_2014.rename(columns={'usertype': 'User_Type'}, inplace=True)
new_df_2014.rename(columns={'birth year': 'Birth_Year'}, inplace=True)
new_df_2014.rename(columns={'gender': 'Gender'}, inplace=True)
new_df_2014.rename(columns={'start station name': 'Start_Station_Name'}, inplace=True)
new_df_2014.rename(columns={'end station name': 'End_Station_Name'}, inplace=True)
new_df_2014.rename(columns={'start station id': 'Start_Station_ID'}, inplace=True)
new_df_2014.rename(columns={'end station id': 'Start_Station_ID'}, inplace=True)
new_df_2014.head(15)
```

```python
# drop any rows with NAN values
new_df_2014.dropna(how='all')
new_df_2014.tail(5)
```

```python
# replace any '\N' values with empty strings, then 
new_df_2014 = new_df_2014.replace(r'\\N','', regex=True)
new_df_2014['Birth_Year'].replace('', pd.np.nan, inplace=True)
new_df_2014.head(15)
```

```python
filter = new_df_2014["Birth_Year"] != ""
final_2014 = new_df_2014[filter]
final_2014.head(15)
```
